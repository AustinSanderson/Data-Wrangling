```python
import pandas as pd
```

```python
#Reading in the spreadsheet
df = pd.read_excel('CoC_Dict + Pov + Incar + Climate + Rents + CannaPol + PopDensity + Net Migration.xlsx')
print(df)
```

```python
# sorting by county FIPS code
df.sort_values("FIPS State County", inplace = True)
print(df)
```

```python
#Dropping rows with duplicate values with specified column subset, keeping the first
df.drop_duplicates(subset =['FIPS State County', 'year', 'state'], keep ="first", inplace = True)
print(df)
```
