```python
import pandas as pd
```

```python
filt_list = ['CoC_number', 'year', 'total_county_pops']
df = pd.read_excel('Recoded - Master - Aggregated Homeless, Beds, Incarceration, Poverty,  Income, Rents, Climate, Election Votes, and Cannabis Policy (2007 - 2020).xlsx', sheet_name='Populations', usecols=filt_list)
df
```

```python
#Creating the list of common group IDs
group_list = ['CoC_number', 'year']

df['total_county_pops'].fillna(df.groupby('CoC_number')['total_county_pops'].transform('mean'), inplace=True)
df['total_county_pops'].fillna(df['total_county_pops'].mean(), inplace=True)
```
